<h3>Check for the Normality, Zero mean, Homoscedasticity</h3>

<h4 style="color:#001e8a">2./ Zero mean: As we have the <b style="color:#ff6363">intercept</b>, so there will be no zero mean </h4>

<h4 style="color:#001e8a">As our observations are not coming from <b style="color:#ff6363">time series</b>, each row contains different customer selling their cars, <br/> <br/> so <b style="color:#ff6363">these observations are not dependent on each other </b></h4>

<h3>Check for Multicollinearity</h3>

```python
data_cleaned.columns
```

<h4 style="color:#001e8a">There might be a relation between <b style="color:#ff6363">Mileage and Year</b> - The newer the car, the lower its mileage</h4>

<h4 style="color:#001e8a">-> <b style="color:#ff6363">sklearn</b> does not have method to check this assumption, so we must turn to <b style="color:#ff6363">statsmodels</b>

<h3>Use variance_inflation_factor (VIF) from stasmodels to check Multicollinearity</h3>

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# define the features we want to check
variables = data_cleaned[['Mileage', 'EngineV', 'Year']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
vif
```

$VIF \in [1,+\inf)$
<p><b>VIF = 1: no multicollinearity <br/>1 &lt; VIF &lt; 5: perfectly okay <br/> 10 &lt; VIF: remove</b></p>

```python
data_no_multicollinearity = data_cleaned.drop(['Year'],axis=1)
data_no_multicollinearity
```

<h3>Create Dummy Variable (Categorical Variable) with pandas.get_dummies</h3>

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first = True)
```

```python
data_with_dummies
```

## Exercise: Dummies and VIF

### Part 1
Calculate the variance inflation factors for all variables contained in data_preprocessed. Anything strange?

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
# define the features we want to check
variables = data_with_dummies[cols]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
vif
```

Obviously, 'log_price' has a very high VIF. This implies it is most definitely **linearly correlated** with all the other variables. And this is no surprise! We are using a linear regression to determine 'log_price' given values of the independent variables! This is exactly what we expect - a linear relationship!

However, to actually assess multicollinearity for the predictors, we have to drop 'log_price'. The multicollinearity assumption refers only to the idea that the **independent variables** shoud not be collinear.

### Part 2
As mentioned in the lecture, your task is to calculate the variance inflation factor (VIF) of all variables including the dummies (but without the dependent variable).

```python
# We will drop the log_price
cols = ['Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
# define the features we want to check
variables = data_with_dummies[cols]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values,i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
vif
```

As you can see, all VIFs are pretty much acceptable. The ones that are particularly high are 'EngineV' and 'Registration_yes'. 

We already discussed 'EngineV' in the lecture.

In the case of <b>registration</b>, the main issue is that **most values are 'yes'** so all types of problems come from there. One way this imbalance manifests is in multicollinearity. Remember that all independent variables are pretty good at determining 'log_price'? Well, if 'registration' is always 'yes', then if we predict 'log_price' we are predicting registration, too (it is going to be 'yes'). That is why, whenever a single category is so predominant, we may just drop the variable. 

Note that it will most probably be insignificant anyways.

### Part 3
Now calculate the VIFs for a data frame where we include the dummies, without 'log_price', but DO NOT DROP THE FIRST DUMMY. Anything strange now?

```python
# To solve this one, we must create a new variable with dummies, without dropping the first one
data_with_dummies_new = pd.get_dummies(data_no_multicollinearity)#, drop_first=True)
data_with_dummies_new.head()
```

```python
# Let's simply drop 'log_price' from this new variable
variables = data_with_dummies_new.drop(['log_price'],axis=1)
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
vif
```

The result that we get is that very interesting indeed. Most VIFs are equal to inf, or plus infinity.

We even got an warning: RuntimeWarning: divide by zero encountered in double_scalars, vif = 1. / (1. - r_squared_i)

The main reason is what we've discussed before. When a car is an 'Audi' all other brand dummies are 0. When a car is not 'Audi', at least one of them will be 1. By including all dummies have introduced multicollinearity (perfect multicollinearity)!!!

If we run a regression including all these dummies, the coefficients would be inflated and completely off-mark.

Now you see why we need to drop one of the dummy variables for each feature.

### Rearrange

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
data_preprocessed = data_with_dummies[cols]
data_preprocessed.describe()
```

## Linear Regression model

### Declare the inputs and the targets

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

### Scale the Data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

### train_test_split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

### Create the regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Targets (y_train)', size=15)
plt.ylabel('Targets (y_train)', size=15)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.displot(y_train-y_hat, kde=True)
plt.title("Residual PDF", size =15)
```

```python
reg.score(x_train, y_train)
```

```python
x_stat = x_train
y_stat = y_train
x_stat = sm.add_constant(x_stat)
model_stat = sm.OLS(y_stat, x_stat).fit()
model_stat.summary()
```

```python
#display adjusted R-squared
1 - (1-reg.score(x_train, y_train))*(len(y_train)-1)/(len(y_train)-x_train.shape[1]-1)
```

Note: <p> If we add an independent variable to a model every time then, the R-squared increases, despite the independent variable being insignificant. </p>
<p>This means that the R-square value increases when an independent variable is added despite its significance.</p>
<p>Adjusted R-squared increases only when the independent variable is significant and affects the dependent variable. </p>
