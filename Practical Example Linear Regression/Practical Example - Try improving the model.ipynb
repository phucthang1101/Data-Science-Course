## Linear Regression model

### Declare the inputs and the targets

```python
targets = data_with_dummies['Price']
inputs = data_with_dummies.drop(['Price'],axis=1)
targets
```

### Scale the Data

<h4 style="color:#001e8a">The method <b style="color:#ff6363">"fit"</b> will calculate <b style="color:#ff6363">the mean and standard deviation </b> of feature "inputs" and these values will be stored in object <b style="color:#ff6363">"scaler"</b></h4>

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

<h4 style="color:#001e8a"><b style="color:#ff6363">StandardScaler.transform</b> will transform (standardize) our input using the information contained in the <b style="color:#ff6363">"scaler"</b> (subtract the mean and divide by standard deviation)</h4>

```python
inputs_scaled = scaler.transform(inputs)
inputs_scaled
```

### train_test_split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)

#check if the x_train extracted from inputs_scaled or not

# type(inputs_scaled[0])
# solution = np.argwhere(inputs_scaled == x_train[3][0])
# print(solution)
#inputs_scaled[0].where(inputs_scaled[0] == x_test[0,0])
y_train.describe()
```

```python
y_test.sort_index()
```

### Create the regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
y_hat_describe = pd.DataFrame(y_hat)
y_hat_describe.describe()
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Targets (y_hat)', size=15)
plt.ylabel('Targets (y_train)', size=15)

plt.show()
```

```python
sns.displot(y_train-y_hat, kde=True)
plt.title("Residual PDF", size =15)
```

```python
reg.score(x_train, y_train)
```

```python
x_stat = x_train
y_stat = y_train
x_stat = sm.add_constant(x_stat)
model_stat = sm.OLS(y_stat, x_stat).fit()
model_stat.summary()
```

### Testing

```python
y_hat_test = reg.predict(x_test)
```

```python

plt.axline([0, 0], [1, 1], color="orange", linewidth = 3,zorder=0)

plt.scatter(y_test,y_hat_test, zorder=2)
plt.xlabel('Targets (y_test)', size=18)
plt.ylabel('Predictions (y_hat_test)', size=18)


plt.show()
```

```python
df_performance = pd.DataFrame(y_hat_test,columns=['Prediction'])
df_performance.head()
```

```python
df_performance['Target'] = y_test
df_performance
```

```python
y_test = y_test.reset_index(drop=True)
# #y_test = y_test.drop(columns=[index])
# y_test = y_test.droplevel(0)
y_test
```

```python
df_performance['Residual'] = df_performance['Target'] - df_performance['Prediction']
df_performance['Difference%'] = np.absolute(df_performance['Residual']/df_performance['Target']*100)
df_performance
```
